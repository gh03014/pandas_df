```python
#저장된 csv파일을 0번째줄을 인덱스 시작으로 읽어온다
data_health = pd.read_csv("health.csv", index_col=0)
data_health
```

```python
#특정 컬럼(NUM_CODE)을 기준으로 오름차순 정렬
data_health = data_health.sort_values(by='NUM_CODE', axis=0)
data_health
```

```python
#특정 컬럼(NUM_CODE)을 기준으로 내림차순 정렬, ascending=False 설정
data_health = data_health.sort_values(by='NUM_CODE', axis=0, ascending=False)
data_health
```

```python
#특정 컬럼(NUM_CODE)삭제
data_health = data_health.drop(['NUM_CODE'], axis=1)
data_health
```

```python
# index 를 기준으로 오름차순 정렬
data_health = data_health.sort_index()
data_health
```

```python
#특정 컬럼('WAIST') 값만 출력
data_health['WAIST']
```

```python
#특정 컬럼에 조건식 설정, 값이 80일때
data_health['WAIST'] == 80
```

```python
#LOC를 이용해 조건식 결과가 TRUE인 경우에만 출력
data_health.loc[data_health['WAIST'] == 80]
```
